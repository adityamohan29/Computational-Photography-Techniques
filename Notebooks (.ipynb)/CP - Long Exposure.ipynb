```python
import cv2
import matplotlib.pyplot as plt
```

```python
def exposure(stream):
    (rAvg, gAvg, bAvg) = (None, None, None)
    total = 0
    while True:
        (grabbed,frame) = stream.read()
    
        if not grabbed:
            break
        
        (B,G,R) = cv2.split(frame.astype("float"))
    
        if rAvg is None:
            rAvg = R
            bAvg = B
            gAvg = G
        
        else:
            rAvg = (rAvg*total + R)/(total+1)
            bAvg = (bAvg*total + B)/(total+1)
            gAvg = (gAvg*total + G)/(total+1)
        
        total = total+1
        
    return rAvg,gAvg,bAvg
```

```python
stream = cv2.VideoCapture("C://Users//Aditya Mohan//Documents//vid1.mp4")
rAvg,gAvg,bAvg = exposure(stream)
```

```python
pic = cv2.merge([rAvg,gAvg,bAvg]).astype("uint8")
fig = plt.figure(figsize=(15, 15))
plt.imshow(pic)
```
